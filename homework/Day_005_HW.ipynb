```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd

# 設定 data_path
dir_data = './data/'
```

```python
f_app_train = os.path.join(dir_data, 'application_train.csv')
app_train = pd.read_csv(f_app_train)
df = pd.DataFrame(app_train)
df.head()
```

```python
# 計算任意欄位的平均數及標準差
t = 'AMT_INCOME_TOTAL'
x = df[t]
mean = df[t].mean()
print(t, 'mean', mean)
std = df[t].std()
print(t, 'std', std)
```

```python
# 畫出任意欄位的直方圖
t = 'AMT_INCOME_TOTAL'
a = df[t]
print('min', a.min())
print('max', a.max())
plt.hist(a, bins = np.linspace(0, 210000, 10)) 
plt.title(t) 
plt.show()
```
